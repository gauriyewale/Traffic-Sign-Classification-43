### Classify 43 different traffic signs using a Deep Neural Network

```python
#Import required libraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle #(Serialization and Deserializatioin of python objects)
import warnings
warnings.filterwarnings('ignore')
```

#### Python pickle module is used for serializing and de-serializing a Python object structure. Any object in Python can be pickled so that it can be saved on disk. What pickle does is that it “serializes” the object first before writing it to file. Pickling is a way to convert a python object (list, dict, etc.) into a character stream. The idea is that this character stream contains all the information necessary to reconstruct the object in another python script

- Pickling: It is a process where a Python object hierarchy is converted into a byte stream.
- Unpickling: It is the inverse of Pickling process where a byte stream is converted into an object hierarchy.

```python
#Read train, test and validation data from pickle files

with open('traffic-signs-data/train.p', mode='rb') as train:
    training_dataset = pickle.load(train)

with open('traffic-signs-data/valid.p', mode='rb') as validation:
    validation_dataset = pickle.load(validation)
    
with open('traffic-signs-data/test.p', mode='rb') as test:
    test_dataset = pickle.load(test)
```

```python
#Create feature and label dataframes for train, validation and test purposes

X_train, y_train = training_dataset['features'], training_dataset['labels']
X_test, y_test = test_dataset['features'], test_dataset['labels']
X_validation, y_validation = validation_dataset['features'], validation_dataset['labels']
```

```python
X_train.shape, y_train.shape
```

```python
plt.imshow(X_train[10])
y_train[10]
```

## Data preprocessing

```python
from sklearn.utils import shuffle

X_train, y_train = shuffle(X_train, y_train)
```

```python
#Convert to grayscale

X_train_gray = np.sum(X_train/3, axis=3, keepdims=True)
X_validation_gray = np.sum(X_validation/3, axis=3, keepdims=True)
X_test_gray = np.sum(X_test/3, axis=3, keepdims=True)
```

```python
X_validation_gray.shape
```

```python
#Normalize grayscaled images

X_train_norm_gray = (X_train_gray - 128)/128
X_validation_norm_gray = (X_validation_gray - 128)/128
X_test_norm_gray = (X_test_gray - 128)/128
```

```python
#Normalized and grayscaled image

plt.imshow(X_train_norm_gray[20].squeeze(), cmap='gray')
```

```python
#Original image

plt.imshow(X_train[20])
```

## Model training

<img src="image.png">

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
leNet_model = Sequential()
leNet_model.add(Conv2D(filters=6, kernel_size=(5,5), activation='relu', input_shape=(32,32,1)))
leNet_model.add(AveragePooling2D())

leNet_model.add(Conv2D(filters=16, kernel_size=(5,5), activation='relu'))
leNet_model.add(AveragePooling2D())

leNet_model.add(Flatten())

leNet_model.add(Dense(units=120, activation='relu'))
leNet_model.add(Dense(units=84, activation='relu'))

leNet_model.add(Dense(units=43, activation='softmax'))
```

```python
leNet_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(lr=0.001), metrics=['accuracy'])
```

```python
history = leNet_model.fit(X_train_norm_gray, y_train, batch_size=512, epochs=10, verbose=1, validation_data=(X_validation_norm_gray, y_validation))
```

## Model Evaluation

```python
score = leNet_model.evaluate(X_test_norm_gray, y_test)
print(f"Test Accuracy: {score[1]}")
```

```python
history.history.keys()
```

```python
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
epochs = range(len(accuracy))

plt.plot(epochs, accuracy, 'b*', label='Training Accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Val Accuracy')
plt.title('Training vs Validation Accuracy')
plt.legend()
```

```python
plt.plot(epochs, loss, 'r*', label='Training Loss')
plt.plot(epochs, val_loss, 'r', label='Val Loss')
plt.title('Training vs Validation Loss')
plt.legend()
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix

y_pred = leNet_model.predict_classes(X_test_norm_gray)
y_true = y_test

cm = confusion_matrix(y_true, y_pred)
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(cm, annot=True)
```

```python
L = 7
W = 7
fig, axes = plt.subplots(L, W, figsize = (15,15))
axes = axes.ravel() #Flatten the array

for i in np.arange(0, L * W):  
    axes[i].imshow(X_test[i])
    axes[i].set_title(f"Prediction={y_pred[i]}\n True={y_true[i]}")
    axes[i].axis('off')

plt.subplots_adjust(wspace=1)
```
